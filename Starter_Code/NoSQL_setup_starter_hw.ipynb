2. Find the BusinessTypeID for "Restaurant/Cafe/Canteen" and return only the `BusinessTypeID` and `BusinessType` fields.

```python
#Cell 12 of homework

# Find the BusinessTypeID for "Restaurant/Cafe/Canteen" and return only the BusinessTypeID and BusinessType fields
query={"BusinessType":"Restaurant/Cafe/Canteen"}
fields={"BusinessTypeID":1,"BusinessType":1}

#results as a list and saved to a variable
results=list(establishments.find(query, fields))

#pretty print results
pprint(results)
```

3. Update the new restaurant with the `BusinessTypeID` you found.

```python
#Cell 13 of homework

# Update the new restaurant with the correct BusinessTypeID
update_query={"BusinessType": "Restaurant/Cafe/Canteen"}
update_values = {"$set": {"BusinessTypeID": 1}}

#update the document
establishments.update_one(update_query, update_values)
```

```python
updated_restaurant = establishments.find_one({"BusinessName": "Penang Flavours"})

if updated_restaurant:
    if "BusinessTypeID" in updated_restaurant:
        pprint(updated_restaurant["BusinessTypeID"])
    else:
        print("BusinessTypeID field is missing in the updated document.")
else:
    print("No document found for BusinessName: Penang Flavours")
```
